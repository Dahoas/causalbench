```python
from causalscbench.utils import retrieve_ensembl_gene_info
ensembl_names = data["var_names"]
ensembl_info = dict()
for i, ensembl_name in enumerate(ensembl_names):
    print(i)
    info = retrieve_ensembl_gene_info(ensembl_name)
    ensembl_info[ensembl_name] = info
```

```python
len(ensembl_info)
print(ensembl_info.keys())
print(ensembl_info["ENSG00000242485"])
```

```python
hgnc_gene_names = []
ensembl_name_to_hgnc = dict()
for name in ensembl_names:
    hgnc_gene_names.append(ensembl_info[name]["display_name"])
    ensembl_name_to_hgnc[name] = ensembl_info[name]["display_name"]
print(len(hgnc_gene_names))
print(hgnc_gene_names[:10])
```

```python
hgnc_interventions = []
new_expression_matrix = []
cnt = 0
for sample, intervention in zip(data['expression_matrix'], data["interventions"]):
    if intervention in ensembl_name_to_hgnc:
        hgnc_interventions.append(ensembl_name_to_hgnc[intervention])
        new_expression_matrix.append(sample.tolist())
        cnt += 1
    elif intervention != "excluded":
        new_expression_matrix.append(sample.tolist())
        hgnc_interventions.append(intervention)
new_expression_matrix = np.array(new_expression_matrix)
```

```python
#print(hgnc_interventions)
print(new_expression_matrix.shape)
print(new_expression_matrix.dtype)
print(len(hgnc_interventions))
print(set(hgnc_interventions))
print("excluded" in hgnc_interventions)
print(cnt)
print("temp")
```

```python
np.savez_compressed("dataset_rpe1_filtered_hgnc.npz", expression_matrix=new_expression_matrix, interventions=hgnc_interventions, var_names=hgnc_gene_names)
```
